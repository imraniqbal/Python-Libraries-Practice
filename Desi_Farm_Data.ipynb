```python
import pandas as pd
df_2019 = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Farm 2019 Expenses vs. Income july to december.xlsx',sheet_name=1)
df_2020a = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Farm Expenses Jan-Jun 2020.xlsx',sheet_name=1)
df_2020b = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Farm Expenses July-Dec 2020.xlsx',sheet_name=1)
df_2021a = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Farm Expenses Jan- June 2021.xlsx',sheet_name=1)
df_2021b = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Farm Expenses July-Dec 2021.xlsx',sheet_name=1)
df_2022a = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Farm Expenses Jan-Jun 2022.xlsx',sheet_name=1)
df_2022b = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Farm Expenses Jul-Dec 2022.xlsx',sheet_name=1)
df_2023a = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Desi Farm Report Jan-Jun 2023.xlsx',sheet_name=1)
df_2023b = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Desi Farm Report July to Dec 2023.xlsx',sheet_name=1)
df_2024a = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Desi Farm Report Jan-June 2024.xlsx',sheet_name=1)
df_2024b = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Desi Farm Report July-Dec 2024.xlsx',sheet_name=1)
df_2025a = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Desi Farm Report Jan-June 2025.xlsx',sheet_name=1)
df_2025b = pd.read_excel('/content/drive/MyDrive/DesiFarmDataExcelforColab/Desi Farm Report July to Dec 2025.xlsx',sheet_name=1)
```

```python
df_2020b_selected_columns = pd.DataFrame(df_2020b.iloc[:,[1,2,3,4]])
df_2020b_renamed_columns = df_2020b_selected_columns.rename(columns={
    'Transactions Details': 'Date',
    'Unnamed: 2': 'Amount(Rs.)',
    'Unnamed: 3': 'Description',
    'Unnamed: 4': 'Category'
})
df_2020b_renamed_columns.iloc[3:].head(5)
```

```python
# function to do all the above tasks to process all files
def process_farm_df(df):
    # Select columns by positional index [1, 2, 3, 4]
    df_selected = df.iloc[:, [1, 2, 3, 4]].copy()

    # Rename columns to standardized names
    df_renamed = df_selected.rename(columns={
        df_selected.columns[0]: 'Date',
        df_selected.columns[1]: 'Amount(Rs.)',
        df_selected.columns[2]: 'Description',
        df_selected.columns[3]: 'Category'
    })

    # Skip top header/blank rows (start from index row 3 onwards)
    df_cleaned = df_renamed.iloc[3:].copy()

    # Remove any rows where Date is null/empty
    df_cleaned = df_cleaned.dropna(subset=['Date'])

    return df_cleaned
```

```python
# List of all your raw DataFrames
raw_dfs = [
    df_2019,
    df_2020a, df_2020b,
    df_2021a, df_2021b,
    df_2022a, df_2022b,
    df_2023a, df_2023b,
    df_2024a, df_2024b,
    df_2025a, df_2025b
]

# Apply the function to each DataFrame
processed_dfs = [process_farm_df(df) for df in raw_dfs]

# Concatenate all processed DataFrames into one master DataFrame
df_all_years = pd.concat(processed_dfs, ignore_index=True)

# Preview the final merged DataFrame
df_all_years
```
