```python
x,y = sp.symbols('x y')
formula = (x+y)**2
expanded_formula = sp.expand(formula)
print(f'Expanded formula: {expanded_formula}')
display(expanded_formula)
```

```python
equation = x**2 + 5*x + 6
factors = sp.factor(equation)
display(factors)
```

```python
import sympy as sp
x= sp.symbols('x')
y= x**3 + 2*x**2 + 5
# Derivative of y with respect to x
dy_dx = sp.diff(y,x)
print('derivative of y with respect to x:')
display (dy_dx)
integral_y =sp.integrate(dy_dx,x)
print('integral of dy/dx with respect to x:')
display(integral_y)
# definite integral of y from 0 to 3
definite_integral_y = sp.integrate(y,(x,0,3))
print('definite integral of y from 0 to 3:')
display(definite_integral_y)
```

```python
import sympy as sp
from IPython.display import display, Math
x = sp.symbols('x')
lim_val = sp.limit(sp.sin(x)/x,x,0)
print('Limit of sin(x)/x as x approaches 0 i.e.')
display(Math(r"\lim_{x \to 0} \frac{\sin x}{x}"))
display(lim_val)
solution =sp.solve(x**2 -5*x +6,x )
print('Solutions to the equation x**2 -5*x +6 = 0:')
display(solution)   
```
